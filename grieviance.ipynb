```python
grv['Product'] = grv['Product'].replace(
    {'Money transfer, virtual currency, or money service': 'Money transfer or virtual currency',
        'Virtual currency': 'Money transfer or virtual currency',
        'Money transfers':'Money transfer or virtual currency'
    })

grv['Product'] = grv['Product'].replace(
    {
        'Payday loan, title loan, or personal loan': 'Payday loan, Title loan, or Personal loan or Advance loan',
        'Payday loan': 'Payday loan, Title loan, or Personal loan or Advance loan',
        'Payday loan, title loan, personal loan, or advance loan':'Payday loan, Title loan, or Personal loan or Advance loan'
    }
)

grv['Product']=grv['Product'].replace(
    {
        'Debt or credit management':'Debt or Credit Management',
        'Debt collection':'Debt or Credit Management'
    }
)

grv['Product']=grv['Product'].replace(
    {
        'Credit reporting':'Credit Reporting or other Personal Consumer Report',
        'Credit reporting or other personal consumer reports':'Credit Reporting or other Personal Consumer Report'
    }
)

grv['Product']=grv['Product'].replace(
    {
        'Credit card or prepaid card':'Credit card or Prepaid card',
        'Prepaid card':'Credit card or Prepaid card'
    }
)

grv['Product']=grv['Product'].replace(
    {
        'Other financial service':'Bank Account or Service',
        'Bank account or service':'Bank Account or Service'
    }
)


product_counts = grv['Product'].value_counts()

print(product_counts)
```

```python
grv.rename(columns={'Sub-issue': 'Sub_issue','Consumer complaint narrative':'Consumer_complaint_narrative'}, inplace=True)
```

```python
grv['Issue']=grv['Issue'].apply(str)
grv['Sub_issue']=grv['Sub_issue'].apply(str)
grv['Consumer_complaint_narrative']=grv['Consumer_complaint_narrative'].apply(str)
```

```python
grv['X'] = grv['Issue'] + " " + grv['Sub_issue'] + " " + grv['Consumer_complaint_narrative']
```

```python
grv['X']
```

```python
grv.rename(columns={'Product':'Y'},inplace = True)
```

```python
grv.X.str.lower()                # make the upper or lower case before building the model
```

```python
from nltk.corpus import stopwords
l1 = list(stopwords.words('english'))
```

```python
import string
string.punctuation
```

```python
def text_process(mess):
    global nopunc
    nopunc = [char for char in mess if char not in string.punctuation ]
    nopunc = "".join(nopunc)

    return[word for word in nopunc.split()if word not in l1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(grv['X'])
```

```python
bow_transformer.vocabulary_
```

```python
TDM = bow_transformer.transform(grv['X'])
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x, train_y, test_y = train_test_split(TDM,grv.Y)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(train_x , train_y)
pred_test = logreg.predict(test_x)
pred_test
from sklearn.metrics import confusion_matrix
tab1 = confusion_matrix(test_y,  pred_test)
tab1
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,pred_test))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train_x, train_y)
pred_nb = nb.predict(test_x)
from sklearn.metrics import confusion_matrix
tab_nb = confusion_matrix(test_y, pred_nb)
tab_nb
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(train_x , train_y)
pred_dt = dt.predict(test_x)
tab_dt = confusion_matrix(test_y , pred_dt)
tab_dt
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,pred_dt))
```
